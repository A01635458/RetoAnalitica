```python
from sklearn.linear_model import LinearRegression as lr
from sklearn.preprocessing import PolynomialFeatures as pf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.impute import SimpleImputer
import seaborn as sns
```

```python
df = pd.read_csv("vgsales.csv")
```

```python
x = df['Year'].to_numpy().reshape((-1,1))
y = df['Global_Sales'].to_numpy()
```

```python
polynomial = pf(degree=2, include_bias=False)
```

```python
polynomial_x = polynomial.fit_transform(x)
```

```python
model_linear = lr()
model_linear.fit(polynomial_x, y)
```

```python
y2 = model_linear.predict(polynomial_x)
```

```python
model_linear.score(polynomial_x,y)
```

```python
plt.scatter(x,y)
plt.plot(x,y2,"r")
plt.xlabel("Año de Lanzamiento")
plt.ylabel("Ventas Globales (MDD)")
plt.show()
```
