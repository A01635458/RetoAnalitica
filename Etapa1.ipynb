```python
import pandas as pd
```

```python
dt = pd.read_csv("dataset.csv", encoding="latin1")
dt.info()
```

```python
print(dt.columns)
```

```python
mean_values = dt.groupby('Programa/Fondo')[' Monto de apoyo ']
print("Agrupamos los programas, y obtuvimos el promedio de monto de apoyo de c/u")
print(mean_values.mean())
```

```python
mean_values.describe()
```

```python
print("Agrupamos los programas, y obtuvimos la mediana del monto de apoyo de c/u")
print(mean_values.median())
```

```python
print("Agrupamos los programas, y obtuvimos la desviación estándar del monto de apoyo de c/u")
print(mean_values.std())
```

```python
print("Programa de apoyo más proporcionado:")
print(dt['Programa/Fondo'].mode())
```

```python
print("Beneficiario más repetido")
print(dt['Nombre del beneficiario'].mode())
```

```python
max1 = max(dt[' Monto de apoyo '])
print("Mayor donación")
print(max1)
```

```python
min1 = min(dt[' Monto de apoyo '])
print("Menor donación")
print(min1)
```
